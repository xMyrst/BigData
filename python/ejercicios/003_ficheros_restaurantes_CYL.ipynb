```python
import matplotlib.pyplot as plt
import numpy as np
%pylab inline

fichero = open('../datos/restaurantes.csv')
bbdd = fichero.readlines()

def porProvincias():
    listaPos = []
    dicProvincias = {}
    dicProvincias2 = {}
    for linea in bbdd:
        lineaSaltos = linea.strip()
        lineaMayus = lineaSaltos.title()
        lineaReplace = lineaMayus.replace('24750', 'León') 
        lineaSeparador = lineaReplace.split(';')     
        listaPos.append(lineaSeparador[2])
        listaPos.append(lineaSeparador[5])
        provincia = lineaSeparador[5]
        if provincia in listaPos:
            dicProvincias[lineaSeparador[5]] = listaPos.count(provincia)
        numRest = list(dicProvincias.values())
        for i in numRest:
            if i != 1:
                dicProvincias2[lineaSeparador[5]] = listaPos.count(provincia)
    return dicProvincias2

dic = porProvincias()

def datosTexto(dic):
    print('\nEste es el nº de restaurantes por provincia en la comunidad de Castilla y León')
    print('------------------------------------------------------------------------------\n')
    for x,y in dic.items():
        print(str(x) + ': ' + str(y))
    return

def datosGrafica(dic):   
    nombre = dic.keys()
    votos = dic.values()           # eje y
    x = np.arange(len(votos))    # eje x
    plt.bar(x, votos)            # pinto diagrama de barras
    plt.xticks(x, nombre, rotation=90);      # etiqueto el eje x con los nombres de las variedades
    return

datosTexto(dic)
datosGrafica(dic)
```

```python
def porMunicipios():
    listaPos = []
    dicMunicipios = {}
    dicMunicipios2 = {}
    for linea in bbdd:
        nSaltos = linea.strip()
        nMayus = nSaltos.title()
        nReplace = nMayus.replace('24750', 'León') 
        nSeparador = nReplace.split(';')     
        listaPos.append(nSeparador[6])
        municipio = nSeparador[6]
        if municipio in listaPos:
            dicMunicipios[nSeparador[6]] = listaPos.count(municipio)
            
        numRest = list(dicMunicipios.values())
        for numero in numRest:
            if numero > 10:
                dicMunicipios2[nSeparador[6]] = listaPos.count(municipio)
    return dicMunicipios2
porMunicipios()
```
